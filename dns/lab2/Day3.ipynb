```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('C:/Users/123/Desktop/Dataset 10.21/cut_dns_categories_names.csv').dropna()
df.loc[df["category_id"] == 'Обувь']
```

```python
encoder = LabelEncoder()
train_descrs = df["description"][::8]
test_descrs = df["description"][1::8]
train_cats = encoder.fit_transform(df["category_id"][::8])
test_cats = encoder.transform(df["category_id"][1::8])
```

```python
count_vec = CountVectorizer(max_features=50)
count_train_vecs = count_vec.fit_transform(train_descrs)
count_test_vecs = count_vec.transform(test_descrs)
count_train_vecs
```

```python
tf_vec = TfidfVectorizer(max_features=50)
tf_train_vecs = tf_vec.fit_transform(train_descrs)
tf_test_vecs = tf_vec.transform(test_descrs)
tf_train_vecs
```
