```python
df1 = pandas.read_csv("C:/Users/123/Desktop/Dataset 10.21/products_dns_sorted.csv")

df2=df1.dropna()
```

```python
testX = df2['description'][::20]
trainX = df2['description'][1::20]
testY = df2['category_id'][::20]
trainY = df2['category_id'][1::20]
```

```python
testY
trainY

```

```python
cnt_vec = CountVectorizer()
trainX = cnt_vec.fit_transform(trainX)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB().fit(trainX,trainY)
```

```python
print("{0:.4f}".format(classifier.score(trainX, trainY)))
```

```python
testX = cnt_vec.transform(testX)

print("{0:.4f}".format(classifier.score(testX, testY)))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec_tree = DecisionTreeClassifier().fit(trainX, trainY)
print("{0:.4f}".format(dec_tree.score(trainX, trainY)))
print("{0:.4f}".format(dec_tree.score(testX, testY)))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
predicted = dec_tree.predict(testX)
```

```python
confusion_matrix(testY, predicted)
```

```python
df_cut = df2[df2.isin(index).values]
df_cut
```

```python
testX_cut = df_cut['description'][::10]
trainX_cut = df_cut['description'][1::10]
testY_cut = df_cut['category_id'][::10]
trainY_cut = df_cut['category_id'][1::10]
```

```python
cnt_vec_cut = CountVectorizer()
trainX_cut = cnt_vec_cut.fit_transform(trainX_cut)

testX_cut = cnt_vec_cut.transform(testX_cut)
```

```python
classifier_cut = MultinomialNB().fit(trainX_cut,trainY_cut)
print("{0:.4f}".format(classifier_cut.score(trainX_cut, trainY_cut)))
print("{0:.4f}".format(classifier_cut.score(testX_cut, testY_cut)))
```

```python
predicted_cut = classifier_cut.predict(testX_cut)
```

```python
confusion_matrix(testY_cut, predicted_cut)
```
